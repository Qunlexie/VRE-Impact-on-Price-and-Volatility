df_detrend.date_time = pd.to_datetime(df_detrend.date_time)

def get_daily(y):
    df_isone = df_detrend[df_detrend['operator']== y]
    df_isone['date_time'] = df_isone.date_time.dt.date
    df_ISONE = df_isone.groupby('date_time').sum()[[ 'vre', 'output']]
    df_ISONE['residual'] = df_isone.groupby('date_time').apply(lambda x: np.average(x.residuals, weights=x.output))
    df_ISONE = df_ISONE.reset_index()
    df_ISONE['vre_frac'] = df_ISONE.vre/df_ISONE.output
    df_ISONE['operator'] = y
    return df_ISONE
